# 1. Order of faces using ISOMAP (50 points)

The objective of this question is to reproduce the ISOMAP algorithm results that we have
seen discussed in lecture as an excercise. The le isomap.mat (or isomap.dat) contains 698
images, corresponding to dierent poses of the same face. Each image is given as a 64  64
luminosity map, hence represented as a vector in R4096. This vector is stored as a row in the
le. [This is one of the datasets used in the original paper for ISOMAP, J.B. Tenenbaum,
V. de Silva, and J.C. Langford, Science 290 (2000) 2319-2323.]

Q1a) (20 points) Choose the Euclidean distance between images (i.e., in this case a distance
in R4096). Construct a similarity graph with vertices corresponding to the images,
and tune the threshold  so that each node has at least 100 neighbors. Visualize the
similarity graph (e.g., plot the adjacency matrix, or visualize the graph and illustrate
a few images corresponds to nodes at dierent parts of the graph; you can be a bit
creative here).

```python

data_matrix = sio.loadmat('isomap.mat')
images = data_matrix['images']
images = images.T
data = pd.DataFrame(images.T)

D_Euclidean = pairwise_distances(images)
m,n = D_Euclidean.shape


epsilon = 0
while epsilon <=2000:
    A_Euclidean =  np.zeros((m,n)) ##Adjacency Matrix
    for i in range(m):
        for j in range(n):
            if (D_Euclidean[i,j] <= epsilon) and (D_Euclidean[i,j] != 0.):
                A_Euclidean[i,j] = 1
    if min((sum(A_Euclidean))) >= 100:
        break
    epsilon = epsilon +1

print("The lowest value of the Treshold Epsilon where each points have 100 neighbours atleast:",epsilon)
print("plotting the adjacency matrix")
plt.figure(figsize=(15,10))
sns.heatmap(A_Euclidean,cmap="GnBu")# visulize the adjancency matrix as a heatmap
plt.title("Heatmap - Adjacency Matrix- Euclidean",fontname="Times New Roman",fontweight="bold",fontsize=18)
plt.show()

cp_A_Euclidean=A_Euclidean.copy()
gpx = nx.from_numpy_matrix(cp_A_Euclidean)
pos=nx.spring_layout(gpx)
val=list(pos.values())
plt.figure(figsize=(16,9))
for i in range(len(val)):
    x = val[i][0]
    y = val[i][1]
    plt.plot(x, y, 'bo',marker="+")
plt.title("Scatter - Adjacency Matrix- Euclidean",fontname="Times New Roman",fontweight="bold",fontsize=18)
plt.show()
```

Q1b) (20 points) Implement the ISOMAP algorithm and apply it to this graph to obtain a
d = 2-dimensional embedding. Present a plot of this embedding. Find three points
that are \close" to each other in the embedding space, and show what they look like.
Do you see any visual similarity among them?

```python

I = np.identity(m)
one_vector = np.ones((m,m))
H_Euclidean = np.cov(D_Euclidean)##I - (1/m)*one_vector
C_Euclidean = (-1/(2*m))*((H_Euclidean.dot(np.square(D_Euclidean))).dot(H_Euclidean))


Eig_val_Euclidean, Eig_vec_Euclidean = np.linalg.eig(C_Euclidean)
Eig_vec_Euclidean = Eig_vec_Euclidean[:,np.argsort(-Eig_val_Euclidean)]
Eig_val_Euclidean = Eig_val_Euclidean[np.argsort(-Eig_val_Euclidean)]
leading_Eig_vec_Euclidean = Eig_vec_Euclidean[:,0:2].real
leading_Eig_val_Euclidean = Eig_val_Euclidean[0:2].real

scaling_matrix = np.zeros((2,2))

for i in range(2):
    for j in range(2):
        if i == j:
            scaling_matrix[i,j] = np.sqrt(Eig_val_Euclidean[i])
            
Z_t_Euclidean = leading_Eig_vec_Euclidean.dot(scaling_matrix)
Z_t_Euclidean = pd.DataFrame(Z_t_Euclidean)
Z_t_Euclidean = Z_t_Euclidean.rename(columns= {0:'x',1:"y"})
Z_t_Euclidean["image_no"] = np.array([i for i in range(data.shape[1])])


plt.figure(figsize=(15,10))
p1 = sns.regplot(data=Z_t_Euclidean, x="x",y="y",fit_reg=False)
image_list= [691,140,154, 406,631,260]

for i in image_list:
    p1.text(Z_t_Euclidean["x"][i],Z_t_Euclidean["y"][i],Z_t_Euclidean["image_no"][i],size='medium', color='black', weight='semibold')

plt.show()
plt.figure(figsize=(15,10))
for i in image_list:
        face_reshape = np.reshape(np.asarray(data.iloc[:,i]), (64,64)).T 
        plt.subplot(2,len(image_list)/2,image_list.index(i)+1)
        plt.title('Eigenface_'+str(i))
        plt.imshow(face_reshape)
        plt.axis('off')

fig = plt.figure()
fig.set_size_inches(10, 10)
ax = fig.add_subplot(111)
ax.set_title('2D Components from Isomap of Facial Images')
ax.set_xlabel('Component: 1')
ax.set_ylabel('Component: 2')

x_size = (max(Z_t_Euclidean['x']) - min(Z_t_Euclidean['x'])) * 0.08
y_size = (max(Z_t_Euclidean['y']) - min(Z_t_Euclidean['y'])) * 0.08
for i in range(40):
    img_num = np.random.randint(0, 600)
    x0 = Z_t_Euclidean.loc[img_num, 'x'] - (x_size / 2.)
    y0 = Z_t_Euclidean.loc[img_num, 'y'] - (y_size / 2.)
    x1 = Z_t_Euclidean.loc[img_num, 'x'] + (x_size / 2.)
    y1 = Z_t_Euclidean.loc[img_num, 'y'] + (y_size / 2.)
    img = np.reshape(np.asarray(data.iloc[:,img_num]), (64,64)).T
    ax.imshow(img, aspect='auto', cmap=plt.cm.gray, 
              interpolation='nearest', zorder=100000, extent=(x0, x1, y0, y1))

ax.scatter(Z_t_Euclidean['x'], Z_t_Euclidean['y'], marker='.',alpha=0.7)
plt.show()
```

# 2. Density estimation: Psychological experiments. (50 points)

Importing Dataset and Converting it into a Numpy Array

```python
if not sys.warnoptions:
    warnings.simplefilter("ignore")
n90pol  = pd.read_csv('n90pol.csv')
amygdala_array = n90pol['amygdala'].as_matrix(columns = None)
acc_array = n90pol['acc'].as_matrix(columns = None)
n90pol_matrix = pd.read_csv('n90pol.csv').to_numpy()
n90pol_matrix = n90pol_matrix[:,0:2]
n90pol_matrix = preprocessing.scale(n90pol_matrix)
```

The data set n90pol.csv contains information on 90 university students who participated in
a psychological experiment designed to look for relationships between the size of dierent
regions of the brain and political views. The variables amygdala and acc indicate the volume
of two particular brain regions known to be involved in emotions and decision-making, the
amygdala and the anterior cingulate cortex; more exactly, these are residuals from the pre-
dicted volume, after adjusting for height, sex, and similar body-type variables. The variable
orientation gives the students' locations on a ve-point scale from 1 (very conservative) to 5
(very liberal).

Q2a) (20 points) Form 2-dimensional histogram for the pairs of variables (amygdala, acc).
Decide on a suitable number of bins so you can see the shape of the distribution clearly.

```python
m,n = n90pol_matrix.shape
h0 = 1.06*np.std(n90pol_matrix[:,0])*(m**(-0.2))
h1 = 1.06*np.std(n90pol_matrix[:,1])*(m**(-0.2))
assert h0==h1
bin_number = round((max(n90pol_matrix[:,0]))-min(n90pol_matrix[:,0]))/h1
print ("The optimal number of bins is:",math.floor(bin_number) )


fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
hist, xedges, yedges = np.histogram2d(amygdala_array, acc_array, bins=math.floor(bin_number))
xpos, ypos = np.meshgrid(xedges[:-1]+xedges[1:], yedges[:-1]+yedges[1:])
xpos = xpos.flatten()/2
ypos = ypos.flatten()/2
zpos = np.zeros_like (xpos)
dx = xedges [1] - xedges [0]
dy = yedges [1] - yedges [0]
dz = hist.flatten()
cmap = cm.get_cmap('jet') # Get desired colormap - you can change this!
max_height = np.max(dz)   # get range of colorbars so we can normalize
min_height = np.min(dz)
# scale each z to [0,1], and get their rgb values
rgba = [cmap((k-min_height)/max_height) for k in dz] 
ax.bar3d(xpos, ypos, zpos, dx, dy, dz,color=rgba, zsort='average' )
ax.set_title("Number of bins: "+str(math.floor(bin_number)))
plt.xlabel("amygdala")
plt.ylabel("acc")
```

Q1c) (10 points) Plot the condition distribution of the volume of the amygdala as a function
of political orientation: p(amygdalajorientation = a), a = 1; : : : ; 5. Do the same for
the volume of the acc. Plot p(accjorientation = a), a = 1; : : : ; 5. You may either use
histogram or KDE to achieve the goal.

```python
# conditional probabilty for "amygdala"
# simple statistics for orientation variable
def gaussian_kernel(x):
    return (1/np.sqrt(2*np.pi))*(np.exp(-x**2/2))

# function to calcute the density of a given new data_point
def KernelDensityEstimator(x, x_all, h):
    ProbablityDensityFunction = 0
    for x_i in x_all:
        a = gaussian_kernel((x_i-x)/h)/h
        ProbablityDensityFunction += a
    return ProbablityDensityFunction/len(x_all)

n90pol["orientation"].describe()

x_min = n90pol["amygdala"].min()*2
x_max = n90pol["amygdala"].max()*2

for i in range(2,6):
    new = n90pol[n90pol["orientation"]==i]["amygdala"]
    ProbablityDensityFunction_all = []
    for n in np.linspace(x_min,x_max,500):
        ProbablityDensityFunction = KernelDensityEstimator(n,new,0.02)
        ProbablityDensityFunction_all.append(ProbablityDensityFunction)
    x = np.column_stack([np.linspace(x_min,x_max,500), np.asarray(ProbablityDensityFunction_all)])
    sns.lineplot(x=x[:,0], y = x[:,1])
plt.legend(labels=['Orientation 2', 'Orientation 3', 'Orientation 4',"Orientation 5"])
```

```python
# conditional probabilty for "acc"
# simple statistics for orientation variable
x_min = n90pol["acc"].min()*2
x_max = n90pol["acc"].max()*2

for i in range(2,6):
    new = n90pol[n90pol["orientation"]==i]["acc"]
    ProbablityDensityFunction_all = []
    for n in np.linspace(x_min,x_max,500):
        ProbablityDensityFunction = KernelDensityEstimator(n,new,0.02)
        ProbablityDensityFunction_all.append(ProbablityDensityFunction)
    x = np.column_stack([np.linspace(x_min,x_max,500), np.asarray(ProbablityDensityFunction_all)])
    sns.lineplot(x=x[:,0], y = x[:,1])
plt.legend(labels=['Orientation 2', 'Orientation 3', 'Orientation 4',"Orientation 5"])
```
