## Predict The Flight Ticket Price Hackathon

Flight ticket prices can be something hard to guess, today we might see a price, check out the price of the same flight tomorrow, it will be a different story. We might have often heard travellers saying that flight ticket prices are so unpredictable. Huh! Here we take on the challenge! As data scientists, we are gonna prove that given the right data anything can be predicted. Here you will be provided with prices of flight tickets for various airlines between the months of March and June of 2019 and between various cities.

Size of training set: 10683 records

Size of test set: 2671 records

FEATURES:
Airline: The name of the airline.

Date_of_Journey: The date of the journey

Source: The source from which the service begins.

Destination: The destination where the service ends.

Route: The route taken by the flight to reach the destination.

Dep_Time: The time when the journey starts from the source.

Arrival_Time: Time of arrival at the destination.

Duration: Total duration of the flight.

Total_Stops: Total stops between the source and destination.

Additional_Info: Additional information about the flight

Price: The price of the ticket

## Feature Engineering

```python
big_df.isnull().sum()
```

```python
big_df['Date_of_Journey'] =  pd.to_datetime(big_df['Date_of_Journey'])
```

```python
big_df['Month'] = big_df['Date_of_Journey'].dt.month
big_df['Day']   = big_df['Date_of_Journey'].dt.day
big_df['Year']  = big_df['Date_of_Journey'].dt.year
```

```python
big_df.head()
```

```python
big_df.dtypes
```

```python
big_df.drop(['Date_of_Journey'], axis = 1, inplace = True)
```

```python
big_df['Arrival_Time'] = big_df['Arrival_Time'].str.split(' ').str[0]
```

```python
big_df['Total_Stops'].value_counts()
```

```python
big_df['Total_Stops'].isnull().sum()
```

```python
big_df.loc[big_df['Total_Stops'].isnull(),:]
```

```python
mode = big_df.loc[(big_df['Source'] == 'Delhi') & (big_df['Destination'] == 'Cochin'),'Total_Stops'].mode()[0]
big_df.loc[(big_df['Source'] == 'Delhi') & (big_df['Destination'] == 'Cochin'),'Total_Stops'].value_counts()
```

Since we have only one null balue for Total Stops for a flight which goes from Delhi to Cochin. There are 4015 flights between delhi and cochin, so i would assume that the nan value can be imputed with 1 stop.

Also all the non-stops can be replaced with 0 stops.

```python
big_df['Total_Stops'].fillna(mode, inplace = True)
```

```python
big_df['Total_Stops'].isnull().sum()
```

```python
big_df['Total_Stops'] = big_df['Total_Stops'].replace('non-stop','0 stop')
```

```python
big_df['Total_Stops'].value_counts()
```

```python
big_df['Stop'] = big_df['Total_Stops'].str.split(' ').str[0]
big_df.head()
```

```python
big_df['Stop'] = big_df['Stop'].astype(int)
```

```python
big_df.dtypes
```

```python
big_df.drop(['Total_Stops'], axis = 1, inplace = True)
big_df.head()
```

```python
big_df['Arrival_Hour'] = big_df['Arrival_Time'].str.split(':').str[0].astype(int)
big_df['Arrival_Minute'] = big_df['Arrival_Time'].str.split(':').str[1].astype(int)
big_df.drop(['Arrival_Time'], axis = 1, inplace = True)
```

```python
big_df['Departure_Hour'] = big_df['Dep_Time'].str.split(':').str[0].astype(int)
big_df['Departure_Minute'] = big_df['Dep_Time'].str.split(':').str[1].astype(int)
big_df.drop(['Dep_Time'], axis = 1, inplace = True)
```

```python
## Duration is not needed
big_df.drop(['Duration'], axis = 1, inplace = True)
```

```python
big_df.head()
```

```python
##Getting the row from the route column with MaxLength

big_df['Route'].unique()
field_length = big_df.Route.astype(str).map(len)
print(big_df.loc[field_length.argmax(), 'Route'])
```

The max length row in the "Route" column has 5 arrows

```python
big_df['Route_1'] = big_df['Route'].str.split('→ ').str[0]
big_df['Route_2'] = big_df['Route'].str.split('→ ').str[1]
big_df['Route_3'] = big_df['Route'].str.split('→ ').str[2]
big_df['Route_4'] = big_df['Route'].str.split('→ ').str[3]
big_df['Route_5'] = big_df['Route'].str.split('→ ').str[4]
big_df.drop(['Route'], axis = 1, inplace = True)
```

```python
big_df.corr()
```

he price column appears to be positively correlated to the stop column after looking at the two figures above. Though it is not heavily correlated, we will impute missing price values with the mean of price grouped by each stop value.

```python
big_df_stop_price = big_df.groupby('Stop', as_index=False)['Price'].mean()
big_df_stop_price
big_df_new = pd.merge(big_df,big_df_stop_price, how = 'left', on = 'Stop')
big_df_new['Price_x'].fillna(big_df_new['Price_y'], inplace=True)
big_df_new.drop(['Price_y'], axis = 1, inplace = True)
big_df_new.rename(columns = {'Price_x' : 'Price'}, inplace = True)
```

```python
big_df = big_df_new.copy()
```

```python
big_df.isnull().sum()
```

```python
big_df['Route_1'].fillna("None", inplace = True)
big_df['Route_2'].fillna("None", inplace = True)
big_df['Route_3'].fillna("None", inplace = True)
big_df['Route_4'].fillna("None", inplace = True)
big_df['Route_5'].fillna("None", inplace = True)
```

```python
big_df.isnull().sum()
```

```python
big_df
```

For those who no the India map and geographic location of each cities, we are able to see that the price value is is proportional to the the distance travelled by the plane. So it is better to label encode the Source and Destination categorical variables so that the ML algorithm will understand that the price value is changing based on label encoded values.

```python
encoder = LabelEncoder()
big_df['Airline'] = encoder.fit_transform(big_df['Airline'])
big_df['Source'] = encoder.fit_transform(big_df['Source'])
big_df['Destination'] = encoder.fit_transform(big_df['Destination'])
big_df['Additional_Info'] = encoder.fit_transform(big_df['Additional_Info'])
big_df['Route_1'] = encoder.fit_transform(big_df['Route_1'])
big_df['Route_2'] = encoder.fit_transform(big_df['Route_2'])
big_df['Route_3'] = encoder.fit_transform(big_df['Route_3'])
big_df['Route_4'] = encoder.fit_transform(big_df['Route_4'])
big_df['Route_5'] = encoder.fit_transform(big_df['Route_5'])
```

```python
df_train = big_df[0:10683]
df_test = big_df[10683:]
```

```python
X = df_train.drop('Price', axis = 1)
y = df_train['Price']
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.3, random_state = 0)
```

```python
model = SelectFromModel(Lasso(alpha = 0.005, random_state = 0))
```

```python
model.fit(X_train,y_train)
```

```python
model.get_support()
```

Looking at the results above, it sounds like Year does is not an important feature and it makes sense as the year has no relationship with the price of a flight

```python
selected_features = X_train.columns[model.get_support()]
selected_features
```

```python
X_train = X_train[selected_features]
X_train
```

```python
X_test = X_test[selected_features]
X_test
```

## Model Development

#### Random Forest Regressor

```python
#Randomized Search CV

# Number of trees in random forest
n_estimators = [int(x) for x in np.linspace(start = 100, stop = 1200, num = 24)]
# Number of features to consider at every split
max_features = ['auto', 'sqrt']
# Maximum number of levels in tree
max_depth = [int(x) for x in np.linspace(5, 30, num = 6)]
# max_depth.append(None)
# Minimum number of samples required to split a node
min_samples_split = [2, 5, 10, 15, 100]
# Minimum number of samples required at each leaf node
min_samples_leaf = [1, 2, 5, 10]
```

```python
# Create the random grid
random_grid = {'n_estimators': n_estimators,
               'max_features': max_features,
               'max_depth': max_depth,
               'min_samples_split': min_samples_split,
               'min_samples_leaf': min_samples_leaf}

print(random_grid)
```

```python
rf = RandomForestRegressor()
```

```python
# Random search of parameters, using 3 fold cross validation, 
# search across 50 different combinations
rf_random = RandomizedSearchCV(estimator = rf, param_distributions = random_grid,scoring='neg_mean_squared_error', n_iter = 50, cv = 5, verbose=2, random_state=42, n_jobs = 1)
```

```python
rf_random.fit(X_train,y_train)
```

```python
rf_random.best_params_
```
