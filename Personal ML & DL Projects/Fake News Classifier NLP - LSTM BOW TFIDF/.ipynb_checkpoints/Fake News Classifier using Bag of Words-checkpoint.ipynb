### Importing Necessary Packages

```python
import pandas as pd
import numpy as np
import os
import re

import matplotlib.pyplot as plt
import seaborn as sns

from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
from nltk.stem import WordNetLemmatizer
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer, HashingVectorizer

from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import confusion_matrix
from sklearn.metrics import roc_auc_score
from sklearn.metrics import f1_score
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report

import warnings
warnings.filterwarnings('ignore')
```

### Importing Data

```python
df = pd.read_csv('train.csv')
df
```

### Feature Engineering

```python
df.shape
```

```python
df.isnull().sum()
```

Since we have a total of 20800 records and approimately only 2000 of them are null/nan, dropping all NaN values will still have 18000 records which is a sizable amount for training our model. 

```python
df = df.dropna()
messages = df.copy()
messages.reset_index(inplace = True)
messages.head()
```

#### Removing special characters, stop words and stemming of title column

```python
ps = PorterStemmer()
corpus = []
for i in range(0, len(messages)):
    review = re.sub('[^a-zA-Z]', ' ', messages['title'][i])
    review = review.lower()
    review = review.split()
    
    review = [ps.stem(word) for word in review if not word in stopwords.words('english')]
    review = ' '.join(review)
    corpus.append(review)
corpus[3]
```

#### Creating the Bag of Words model

```python
cv = CountVectorizer(max_features=5000,ngram_range=(1,3))
X = cv.fit_transform(corpus).toarray()
X = pd.DataFrame(X, columns = cv.get_feature_names())
X
```

```python
cv.get_feature_names()[:20]
```

```python
y = messages['label']
y = pd.DataFrame(y)
y['label'] = y['label'].map({0:'Real', 1:'Fake'})
y
```

### Building Models

#### Train Test Split

```python
## Divide the dataset into Train and Test
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=0)
```

#### MultinomialNB Algorithm

```python
classifier=MultinomialNB()
classifier.fit(X_train, y_train)
y_pred = classifier.predict(X_test)
y_pred
```

```python
print("Accuracy:",classifier.score(X_test, y_test))
print(classification_report(y_test, y_pred))

labels = list(set(y_train['label']))
labels.sort()
cm = confusion_matrix(y_test,y_pred, labels = labels)
ax= plt.subplot()
sns.heatmap(cm, annot=True, ax = ax,cmap='Blues',fmt="d"); #annot=True to annotate cells
# labels, title and ticks
ax.set_xlabel('Predicted');ax.set_ylabel('Actual'); 
ax.set_title('Confusion matrix of the MultinomialNB'); 
ax.xaxis.set_ticklabels(labels,rotation = 90); 
ax.yaxis.set_ticklabels(labels,rotation = 0);
```
