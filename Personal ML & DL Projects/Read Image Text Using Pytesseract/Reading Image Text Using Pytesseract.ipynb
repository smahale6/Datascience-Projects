### Importing Necessary Packages

```python
import pytesseract
pytesseract.pytesseract.tesseract_cmd = r'C:\\Users\\D100793\\AppData\\Local\\Programs\\Tesseract-OCR\\tesseract.exe'
import cv2
import PIL.Image as Image
```

### Importing an Image - 1.png

```python
IMAGE_SHAPE = (300, 300)
Image1 = Image.open("1.png").resize(IMAGE_SHAPE)
Image1
```

```python
img = cv2.imread('1.png')
cv2.imshow('sample image',img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

#### Using Pytesseract

```python
text = pytesseract.image_to_string(img)
print(text)
```
